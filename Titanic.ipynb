```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")
titanic.drop(['embark_town','adult_male','sex','alive','class'],axis=1, inplace=True)
titanic['age'].fillna(titanic['age'].mean(),inplace=True)
new_titanic=pd.get_dummies(titanic)
new_titanic['alone']=new_titanic['alone'].replace({True:1, False:0})
new_titanic
```

```python
from sklearn.preprocessing import StandardScaler

norm_scaler = StandardScaler()
new_titanic['age'] = norm_scaler.fit_transform(new_titanic[['age']])
new_titanic['fare'] = norm_scaler.fit_transform(new_titanic[['fare']])
new_titanic
```

```python
import sklearn
from sklearn.model_selection import train_test_split
X = new_titanic.drop(['survived'], axis=1)
y = new_titanic['survived']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=42)
y_train
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver='liblinear', max_iter=1000)
model.fit(X_train, y_train)

y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix

cm_train = confusion_matrix(y_train, y_train_pred)
cm_train
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)
print("Train accuracy of the model is %.2f percent"%(train_accuracy*100))
print("Test accuracy of the model is %.2f percent"%(test_accuracy*100))

train_precision = precision_score(y_train, y_train_pred)
test_precision = precision_score(y_test, y_test_pred)
print("Train precision: %.2f"%(train_precision*100))
print("Test precision: %.2f"%(test_precision*100))

train_recall = recall_score(y_train, y_train_pred)
test_recall = recall_score(y_test, y_test_pred)
print("Train recall: %.2f"%(train_recall*100))
print("Test recall: %.2f"%(test_recall*100))

train_f1 = f1_score(y_train, y_train_pred)
test_f1 = f1_score(y_test, y_test_pred)
print("Train f1: %.2f"%(train_f1*100))
print("Test f1: %.2f"%(test_f1*100))

train_roc_auc = roc_auc_score(y_train, y_train_pred)
test_roc_auc = roc_auc_score(y_test, y_test_pred)
print("Train roc_auc: %.2f"%(train_roc_auc*100))
print("Test roc_auc: %.2f"%(test_roc_auc*100))
```
